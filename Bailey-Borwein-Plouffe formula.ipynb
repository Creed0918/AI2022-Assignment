```python
from decimal import Decimal
from decimal import getcontext
import numpy as np
import pandas as pd
from mpmath import mp
import matplotlib.pyplot as plt

mp.dps = 80
getcontext().prec=80
def BBP(N):
    
    return sum(1/Decimal(16)**k *
        (Decimal(4)/(8*k+1) -
         Decimal(2)/(8*k+4) -
         Decimal(1)/(8*k+5) -
         Decimal(1)/(8*k+6)) for k in range (N))



bbp=[]
bbp_diff=[]
for i in range(1,32):
    bbp.append(BBP(i))
    bbp_diff.append(mp.pi-abs(BBP(i)))




d = {'Pi of Bailey-Borwein-Plouffe formula' : pd.Series(bbp), 'Absolute error' : pd.Series(bbp_diff)}
df = pd.DataFrame(d)
df[:]
```

```python
plt.semilogy(bbp_diff[0:31])
plt.title('Approximation Error of Bailey-Borwein-Plouffe formula')
plt.xlabel('N')
plt.ylabel('Absolute Error')
plt.grid(True)
plt.xlim(0, 31)

plt.show()

```
